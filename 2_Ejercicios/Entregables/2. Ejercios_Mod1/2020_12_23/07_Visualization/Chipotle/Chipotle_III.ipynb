### Step 5. Create a histogram of the top 5 items bought

```python
# 1 mtodo para obtener serie
gr1 = chipo.groupby('item_name').sum().sort_values('quantity', ascending=False).quantity.head(5).iloc[::-1]
#plt.plot(gr1)
#print(gr1)

# 2 forma de sacar los datos
gr1 = chipo.item_name.value_counts().head(5).iloc[::-1]
plt.bar(gr1.index, gr1.values, width=0.5)
plt.xticks(rotation='vertical')
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most Ordered Chipotle\'s items')
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
def remove_dollar(x):
    if '$' in x:
        return float(x.strip('$'))
    
chipo = chip.copy()
chipo['item_price'] = chipo['item_price'].apply(remove_dollar)
```

```python
gr2_x = chipo.groupby('order_id').sum().item_price
gr2_y = chipo.groupby('order_id').sum().quantity
plt.scatter(x=gr2_x, y=gr2_y, color='green')
plt.xlabel('Order Price')
plt.ylabel('Items Ordered')
plt.title('Number of items ordered per order price')
```


- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
import random

num_ale = list(str(random.randint(1000,9999)))

while True:
    entrada = input('Inserta un número de 4 dígitos/ STOP para salir:')
    if entrada.upper() == 'STOP':
        print('Juego finalizado')
        break
    if len(entrada) > 4 or len(entrada) < 3:
        print('El número debe contener 4 dígitos')
        continue
    elif not entrada.isnumeric():
        print('La entrada debe ser un número de 4 dígitos')
        continue
    cows = 0
    bulls = 0
    for pos, num in enumerate(entrada):
        if num == num_ale[pos]:
            cows += 1
        elif num in num_ale:
            bulls += 1
    print('Tu número ' + entrada, ' --->', 'numéro cows = ' + str(cows), 'numéro bulls= ' +str(bulls))
    if cows == 4:
        print('Enhorabuena has ganado')
        resp = input('¿Quieres seguir jugando? (yes/no):').lower()
        while resp != 'yes' or resp != 'no':
            print('Indica correctamente si quieres continuar')
            resp = input('¿Quieres seguir jugando? (yes/no):').lower()
        if resp == 'no':
                print('Fin del juego')
                break
        elif resp == 'yes':
            num_ale = str(random.randint(1000,9999))
 
```

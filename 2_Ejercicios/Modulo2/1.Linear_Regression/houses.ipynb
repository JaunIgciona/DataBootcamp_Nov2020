### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
import pandas as pd

df = pd.read_csv("data/USA_Housing.csv")
df
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
x = df.iloc[:,0:5]
y = df['Price'].values.reshape(-1,1)
```

```python
for col in x:
    X = x[col].values.reshape(-1,1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,  random_state=0)
    model = LinearRegression()
    model.fit(X_train, y_train)
    print(f'--  {col}')
    print('Train score', model.score(X_train,y_train))
    print('Test score', model.score(X_test,y_test))
    y_test_pred = model.predict(X_test)
    plt.scatter(X_test, y_test, color='r')
    plt.plot(X_test, y_test_pred, color='b')
    plt.xlabel(f'{col}')
    plt.xlabel('Price')
    plt.show()
```

```python
# Score for model with all variables

X = x.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,  random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
print('Train score', model.score(X_train,y_train))
print('Test score', model.score(X_test,y_test))
```
